**Scatterplots**

```python
x=np.array(df['GPA'])
y=np.array(df['Hours_of_Sleep'])
plt.plot(x,y,'o') 
A = np.transpose(np.concatenate(([x], np.ones((1, len(x))))))
coords = np.linalg.inv(np.transpose(A) @ A) @ np.transpose(A) @ y
yhat = A @ coords
plt.title('Hours of Sleep vs GPA')
plt.plot(x,y,'o')
plt.plot(x,yhat)
```

**Chi-Square Tests**

```python
import numpy as np
import scipy.stats

# Sleep and GPA
gpa = pd.crosstab(df["Hours_of_Sleep"], df["GPA"])
_, p1, _, _ = scipy.stats.chi2_contingency(gpa)
print('sleep and GPA p-value: ', p1)

# Sleep and Exercise
exercise = pd.crosstab(df["Hours_of_Sleep"], df["Hours_of_Excercise"])
_, p2, _, _ = scipy.stats.chi2_contingency(exercise)
print('sleep and exercise p-value: ', p2)

# Sleep and Extracurricular
extra = pd.crosstab(df["Hours_of_Sleep"], df["Extracurricular"])
_, p3, _, _ = scipy.stats.chi2_contingency(extra)
print('sleep and extracurriculars p-value: ', p3)

# Sleep and Homework
hw = pd.crosstab(df["Hours_of_Sleep"], df["Homework"])
_, p4, _, _ = scipy.stats.chi2_contingency(hw)
print('sleep and homework p-value: ', p4)

# Sleep and Productivity(Exercise, Extracurriculars, and Homework combined)
sum_column = df["Hours_of_Excercise"] + df["Extracurricular"] + df["Homework"]
df['Productivity'] = sum_column
prod = pd.crosstab(df["Hours_of_Sleep"], df["Productivity"])
_, p5, _, _ = scipy.stats.chi2_contingency(prod)
print('sleep and total productivity p-value: ', p5)
```
